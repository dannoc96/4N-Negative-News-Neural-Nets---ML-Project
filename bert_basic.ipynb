```
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
```

```
train_non_adverse = pd.read_csv("nam.csv")
train_adverse = pd.read_csv("am.csv")
train = pd.concat([train_non_adverse, train_adverse], ignore_index=True)
train_nam = train[train.label == "nam"]
train_am = train[train.label == "am"]
train_random = train[train.label == "random"]
train_filtered = pd.concat([train_nam, train_am, train_random], ignore_index=True)
train_filtered = train_filtered.dropna(subset=['title', 'article'])
train_data = pd.DataFrame([])
train_data['text'] = train_filtered[['title', 'article']].agg(' '.join, axis=1)
train_data['label'] = train_filtered.label.map(dict(am=1, nam=0, random=0))
```

```
y_train, y_test = train_test_split(train_data['label'], test_size=0.15, random_state=1)
xtrainfile = "xtrainfile.pkl"
xtestfile = "xtestfile.pkl"
import pickle 
open_file = open(xtrainfile, "rb")
x_train = pickle.load(open_file)
open_file.close() 

open_file = open(xtestfile, "rb")
x_test = pickle.load(open_file)
open_file.close()

y_train = pd.concat([y_train,y_test])
x_train=x_train+x_test
len(x_train)
```

```
len(y_train)
```

```
from simpletransformers.classification import ClassificationModel, ClassificationArgs
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import pandas as pd
train_data = pd.DataFrame({'text':x_train, 'label':y_train})
n=7
seed=42
model_args= ClassificationArgs()
model_args.num_train_epochs=5
model_args.learning_rate = 2e-5
model_args.overwrite_output_dir = True
model_args.max_seq_length = 256
model_args.max_grad_norm = 1
kf = KFold(n_splits=n, random_state=seed, shuffle=True)
results, acc, prec, rec, f1 = [], [],[],[],[]
for train_index, val_index in kf.split(train_data): 
        train_df = train_data.iloc[train_index]
        val_df = train_data.iloc[val_index]
        model = ClassificationModel('bert', 'bert-base-uncased',args=model_args)
        model.train_model(train_df)
        result, model_outputs, wrong_predictions = model.eval_model(val_df, acc=accuracy_score) 
        model_outputs = [np.argmax(i) for i in model_outputs]
        acc.append(accuracy_score(val_df.label, model_outputs))
        prec.append(precision_score(val_df.label, model_outputs))
        rec.append(recall_score(val_df.label, model_outputs))
        f1.append(f1_score(val_df.label, model_outputs))
```

```
print('Accuracy score: ', np.mean(acc))
print('Precision score: ', np.mean(prec))
print('Recall score: ', np.mean(rec))
print('F1 score: ', np.mean(f1))
```
